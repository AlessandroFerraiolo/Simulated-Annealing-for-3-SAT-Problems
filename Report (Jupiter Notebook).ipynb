### Collapsing curves

```python
fig, ax = plt.subplots(figsize = (10, 5))

ax.plot(np.array(list(S2P200.keys())) /200, S2P200.values(), color = 'navy', label = 'N = 200')
ax.plot(np.array(list(S2P300.keys())) /300, S2P300.values(), color = 'darkgreen', label = 'N = 300')
ax.plot(np.array(list(S2P400.keys())) /400, S2P400.values(), color = 'blueviolet', label = 'N = 400')
ax.plot(np.array(list(S2P500.keys())) /500, S2P500.values(), color = 'indigo', label = 'N = 500')
ax.plot(np.array(list(S2P600.keys())) /600, S2P600.values(), color = 'c', label = 'N = 600')

ax.set_xlim(1.5,6)
ax.legend(loc = 'best')
ax.set_title('Collapsed Curves', fontsize=12, fontweight="bold")
ax.set_xlabel('M/N', fontsize=12)
ax.set_ylabel('Percentage (%)', fontsize=12)
ax.grid(alpha=0.5)
```
